## 1 Задача 1.3.12
Знайти ймовiрнiсть того, що серед трьох цифр, кожна з яких вибрана навмання, буде лише 1, 2, 3 рiзних.

Обчисліть теоретичну (повним перебором) та еміричну (симулюванням $100000$ експериментів) імовірності.

```python
import random
from itertools import product

def theoretical():
    total = 10**3

    count_1 = 0
    count_2 = 0
    count_3 = 0

    for d1, d2, d3 in product(range(10), repeat=3):
        s = {d1, d2, d3}
        k = len(s)

        if k == 1:
            count_1 += 1
        elif k == 2:
            count_2 += 1
        elif k == 3:
            count_3 += 1

    p1 = count_1 / total
    p2 = count_2 / total
    p3 = count_3 / total

    print("Теоретичні ймовірності:")
    print(f"  рівно 1 різне значення: {p1:.5f}")
    print(f"  рівно 2 різних значення: {p2:.5f}")
    print(f"  рівно 3 різних значення: {p3:.5f}")
    print(f"  Перевірка суми: {p1 + p2 + p3:.5f}")

def empirical(n=100_000):
    count_1 = 0
    count_2 = 0
    count_3 = 0

    for _ in range(n):
        d1 = random.randint(0,9)
        d2 = random.randint(0,9)
        d3 = random.randint(0,9)

        k = len({d1,d2,d3})
        if k == 1:
            count_1 += 1
        elif k == 2:
            count_2 += 1
        else: # k == 3
            count_3 += 1

    p1 = count_1 / n
    p2 = count_2 / n
    p3 = count_3 / n

    print(f"\nЕмпіричні ймовірності (на основі {n} експериментів):")
    print(f"  рівно 1 різне значення: {p1:.5f}")
    print(f"  рівно 2 різних значення: {p2:.5f}")
    print(f"  рівно 3 різних значення: {p3:.5f}")
    print(f"  Перевірка суми: {p1 + p2 + p3:.5f}")

def main():
    theoretical()
    empirical(100_000)

if __name__ == "__main__":
    main()

```

# 2 Задача 1.3.14
З послiдовностi чисел $1, 2, . . . , n$ вибирають навмання $k$ рiзних чисел.
Яка ймовiрнiсть того, що:
1. кожне з вибраних чисел кратне даному числу $p$;
2. кожне з вибраних чисел кратне хоча б одному з двох взаємно простих чисел $p$ i $q$;
3. серед вибраних чисел є хоча б одне кратне $p$?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k, p, q$.
Виведіть результат для
- $n = 25, k = 5, p = 3, q = 4$;
- $n = 25, k = 10, p = 3, q = 4$.

```python
import math
import random
from itertools import combinations

def is_prime(x):
    if x < 2:
        return False
    for d in range(2, int(math.isqrt(x))+1):
        if x % d == 0:
            return False
    return True

def all_divisible_by_p(combo, p):
    """Подія A: всі числа у вибірці combo кратні p."""
    return all(x % p == 0 for x in combo)

def exactly_m_primes(combo, m):
    """Подія B: рівно m простих чисел серед обраних."""
    count_primes = sum(is_prime(x) for x in combo)
    return (count_primes == m)

def has_no_multiples_of_q(combo, q):
    """Подія C: серед обраних НЕМАЄ жодного числа, кратного q."""
    return all(x % q != 0 for x in combo)


def theoretical(n, k, p=3, m=2, q=5):

    from math import comb

    total = comb(n, k)  # загальна кількість комбінацій
    countA = 0
    countB = 0
    countC = 0

    for combo in combinations(range(1, n+1), k):
        if all_divisible_by_p(combo, p):
            countA += 1
        if exactly_m_primes(combo, m):
            countB += 1
        if has_no_multiples_of_q(combo, q):
            countC += 1

    pA = countA / total
    pB = countB / total
    pC = countC / total

    print("=== Теоретичні ймовірності ===")
    print(f"Подія A (усі кратні {p}): {pA:.5f}")
    print(f"Подія B (рівно {m} простих): {pB:.5f}")
    print(f"Подія C (жодного кратного {q}): {pC:.5f}")
    print(f"Перевірка суми (не обов'язково 1): {pA + pB + pC:.5f}")


def empirical(n, k, trials=100_000, p=3, m=2, q=5):
    """
    Емпірична оцінка ймовірностей подій A, B, C за допомогою випадкового вибору k різних чисел із {1..n}.
    """
    countA = 0
    countB = 0
    countC = 0

    for _ in range(trials):
        combo = random.sample(range(1, n+1), k)

        if all_divisible_by_p(combo, p):
            countA += 1
        if exactly_m_primes(combo, m):
            countB += 1
        if has_no_multiples_of_q(combo, q):
            countC += 1

    pA = countA / trials
    pB = countB / trials
    pC = countC / trials

    print("\n=== Емпіричні ймовірності ===")
    print(f"Подія A (усі кратні {p}): {pA:.5f}")
    print(f"Подія B (рівно {m} простих): {pB:.5f}")
    print(f"Подія C (жодного кратного {q}): {pC:.5f}")
    print(f"Перевірка суми (не обов'язково 1): {pA + pB + pC:.5f}")


def main():
    n = 25
    k = 5
    p = 3
    m = 2
    q = 5

    theoretical(n, k, p, m, q)
    empirical(n, k, 100_000, p, m, q)

if __name__ == "__main__":
    main()


```

# 3 Задача 1.4.6
В коморi знаходяться $n$ пар черевикiв.
З них випадковим чином вибираються $2k$ черевикiв.
Яка ймовiрнiсть того, що серед вибраних черевикiв:
1. вiдсутнi парнi;
2. є рiвно одна комплектна пара;
3. є двi комплектнi пари?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k$.
Виведіть результат для
- $n = 8, k = 4$;
- $n = 13, k = 5$.

```python
import math
import random
from itertools import combinations

def count_pairs(chosen, n):
    pair_count = 0
    freq = [0] * n
    for sock in chosen:
        freq[sock // 2] += 1
    for i in range(n):
        if freq[i] >= 2:
            pair_count += 1
    return pair_count

def theoretical(n, k):
    total = math.comb(2*n, 2*k)
    count_0 = count_1 = count_2 = 0
    for combo in combinations(range(2*n), 2*k):
        p = count_pairs(combo, n)
        if p == 0:
            count_0 += 1
        elif p == 1:
            count_1 += 1
        elif p == 2:
            count_2 += 1
    p0 = count_0 / total
    p1 = count_1 / total
    p2 = count_2 / total
    print("=== Теоретичні ймовірності ===")
    print(f"n={n}, k={k}, загальна кількість = C({2*n},{2*k}) = {total}")
    print(f"Жодної пари (p=0): {p0:.6f}")
    print(f"Рівно 1 пара  (p=1): {p1:.6f}")
    print(f"Рівно 2 пари (p=2): {p2:.6f}")
    print(f"Перевірка суми: {p0+p1+p2:.6f}")

def empirical(n, k, trials=100000):
    count_0 = count_1 = count_2 = 0
    socks = list(range(2*n))
    for _ in range(trials):
        chosen = random.sample(socks, 2*k)
        p = count_pairs(chosen, n)
        if p == 0:
            count_0 += 1
        elif p == 1:
            count_1 += 1
        elif p == 2:
            count_2 += 1
    p0 = count_0 / trials
    p1 = count_1 / trials
    p2 = count_2 / trials
    print("\n=== Емпіричні ймовірності ===")
    print(f"n={n}, k={k}, trials={trials}")
    print(f"Жодної пари: {p0:.6f}")
    print(f"Рівно 1 пара: {p1:.6f}")
    print(f"Рівно 2 пари: {p2:.6f}")
    print(f"Сума: {p0+p1+p2:.6f}")

def main():
    n = 5
    k = 2
    theoretical(n, k)
    empirical(n, k, 100000)
    print("\n--- n=3, k=2 ---")
    theoretical(3, 2)
    empirical(3, 2, 100000)
    print("\n--- n=8, k=4 ---")
    theoretical(8, 4)
    empirical(8, 4, 100000)

if __name__ == "__main__":
    main()
```

# 4 Задача 1.2.15
Нехай $\Omega = {1, 2, \ldots, 2n}$.
Всiм числам приписанi ймовiрностi, пропорцiйнi логарифмам цих чисел.
Знайти цi ймовiрностi.
Знайти ймовiрнiсть того, що в результатi експерименту з’явиться:
1. парне число;
2. непарне число.

Напишіть функцію для обрахунку еміричної (симулюванням $100000$ експериментів) імовірності в залежності від параметра $n$.
Виведіть результат для
- $n = 10$;
- $n = 25$.

```python
import math
import random

def theoretical(n):
    s = sum(math.log(i) for i in range(1, 2*n+1))
    even_sum = sum(math.log(i) for i in range(2, 2*n+1, 2))
    odd_sum = sum(math.log(i) for i in range(1, 2*n+1, 2))
    p_even = even_sum / s
    p_odd = odd_sum / s
    print("Теоретичні ймовірності:")
    print(f"  P(парне) = {p_even:.6f}")
    print(f"  P(непарне) = {p_odd:.6f}")
    print(f"  Перевірка суми: {p_even + p_odd:.6f}")

def empirical(n, trials=100000):
    w = []
    s = sum(math.log(i) for i in range(1, 2*n+1))
    for i in range(1, 2*n+1):
        w.append(math.log(i)/s)
    ev = 0
    for _ in range(trials):
        r = random.random()
        cum = 0
        for i, prob in enumerate(w, start=1):
            cum += prob
            if r < cum:
                if i % 2 == 0:
                    ev += 1
                break
    p_even = ev / trials
    p_odd = 1 - p_even
    print("\nЕмпіричні ймовірності:")
    print(f"  P(парне) = {p_even:.6f}")
    print(f"  P(непарне) = {p_odd:.6f}")
    print(f"  Сума: {p_even + p_odd:.6f}")

def main():
    n = 10
    theoretical(n)
    empirical(n, 100000)
    print("\n--- n=25 ---")
    n = 25
    theoretical(n)
    empirical(n, 100000)

if __name__ == "__main__":
    main()
```
